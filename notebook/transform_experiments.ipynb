01_「会社名」項目の確認

```python
import pandas as pd

# スクレイピング結果csvを読み込んで、「会社名」データ確認
df = pd.read_csv("/app/data/rikunabi.csv")
```

```python
# 会社名の表記確認
print(f"先頭20件:\n{df['会社名'].head(20)}")
print("------------------------------------------------------")
print(f"会社名の出現回数:\n{df['会社名'].value_counts().head(30)}")
print("------------------------------------------------------")
print(f"末尾20件:\n{df['会社名'].tail(20)}")
```

```python
# ノイズや余計な情報の混入
print(
    f"ノイズ混入件数: {df['会社名'].str.contains('本社|支店|本店|営業所| 東京| 大阪| 名古屋', na=False).sum()}"
)

# ノイズや余計な情報の混入確認
df[df["会社名"].str.contains("本社|支店|本店|営業所| 東京| 大阪| 名古屋", na=False)]
```

```python
# 欠損値・異常値・空文字の確認
print(f"欠損値の件数: {df['会社名'].isna().sum()}")
print(f"空文字の件数: {df['会社名'].str.strip().eq('').sum()}")
print(
    f"異常値の件数: {
        df['会社名'].str.contains(r'^[0-9]|^[A-Za-z]', regex=True, na=False).sum()
    }"
)
```

```python
# 異常値の確認
df[df["会社名"].str.contains(r"^[0-9]|^[A-Za-z]", regex=True, na=False)]
```

```python
# 省略表記の確認
print(
    f"省略表記の件数: {df['会社名'].str.contains('(株)|(有)|㍿|（株）|（有）', regex=True, na=False).sum()}"
)
```
